```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#load csv file
df = pd.read_csv("sales_data_10000.csv")
df.head()
```

```python
# average sales colom 
avg_sales = df["Sales"].mean()
print("AVERAGE_SALES:",avg_sales)
```

```python
#visulazation  sales and category by bar chart
df.groupby("Category")["Sales"].sum().plot(kind="bar", color= "skyblue")
plt.title("sales according to category")
plt.xlabel("category")
plt.ylabel("totale_sales")
plt.show()
```

```python
#relation betveen sales and profit by scatter plot
plt.scatter(df['Sales'],df['Profit'], color = "red")
plt.title("relation betveen sales and profit")
plt.xlabel("Sales")
plt.ylabel("profit")
plt.show()
```

```python
#corelation between diffrent catogorys
corelation = df.select_dtypes(include=["number"]).corr()

plt.figure(figsize=(8,6))
sns.heatmap(corelation, annot=True, cmap= "coolwarm", fmt = ".2f")
plt.title("corelation between data")
plt.show()
```
